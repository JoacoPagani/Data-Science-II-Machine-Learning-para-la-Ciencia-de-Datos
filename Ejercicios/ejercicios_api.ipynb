## Ejercicio 1: Consultar una API y Analizar Datos con Pandas


Utiliza la API de JSONPlaceholder (https://jsonplaceholder.typicode.com/) para obtener datos de usuarios (/users) y publicaciones (/posts).

Almacena los datos obtenidos en estructuras de datos de Pandas (por ejemplo, DataFrames).

Combina la información de usuarios y publicaciones en un solo DataFrame utilizando la clave externa (userId).

Realiza algunas operaciones básicas de exploración y análisis de datos con Pandas, como contar el número de publicaciones por usuario, encontrar el usuario con más publicaciones, etc.


```python
#Consulta a Api + Merge entre 2 Df

import requests
import pandas as pd
import json

urlPosts = 'https://jsonplaceholder.typicode.com/posts'
urlUsers = 'https://jsonplaceholder.typicode.com/users'


responsePosts = requests.get(urlPosts)
responseUsers = requests.get(urlUsers)

dataPosts = responsePosts.json()
dataUsers = responseUsers.json()

dfPosts = pd.DataFrame(dataPosts)
dfUsers = pd.DataFrame(dataUsers)

dfMerge = pd.merge(dfPosts, dfUsers, left_on='userId', right_on='id', suffixes=('_post', '_user'))
dfMerge.head()
```

```python
#Analisis de Datos

post_counts = dfMerge.groupby('name')['id_post'].count()
print("Número de publicaciones por usuario:\n", post_counts)

userMasPosts = post_counts.idxmax()
print("\nUsuario con más publicaciones:", userMasPosts)

```
